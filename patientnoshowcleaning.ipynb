# Steps to clean the data
1. The neighbourhoods' names are all capitalized and it doesn't look very nice. Let's fix that =-).
2. When read from the .csv file, the patient ID uses 'e' to show large number. Let's reforamt it as an integer.
3. `[ScheduledDay]` and `[ApponintmentDay]` are not formatted correctly. let's turn them into `datetime` format.
4. Let's also extract the time from `[ScheduledDay]` and put it in a new column.
5. Next, let's extract the date from `[ScheduledDay]` and `[AppointmentDay]`.
6. Transform some columns to category for optimization and ease-of-use.
7. Finally let's save the finished work!

```python
newdata = pd.read_csv("patientnoshow.csv")
newdata['Neighbourhood'] = newdata['Neighbourhood'].str.title()
newdata['PatientId'] = newdata['PatientId'].astype('int64')
newdata["ScheduledDay"] = pd.to_datetime(newdata["ScheduledDay"])
newdata["AppointmentDay"] = pd.to_datetime(newdata["AppointmentDay"])
newdata.insert(4, "ScheduledTime", newdata["ScheduledDay"].dt.time)
newdata["ScheduledDay"] = newdata["ScheduledDay"].dt.date
newdata["AppointmentDay"] = newdata["AppointmentDay"].dt.date
newdata["Scholarship"].astype('category')
newdata["Hipertension"].astype('category')
newdata["Diabetes"].astype('category')
newdata["Alcoholism"].astype('category')
newdata["Handcap"].astype('category')
newdata["SMS_received"].astype('category')
newdata["No-show"].astype('category')
newdata
```

```python
newdata.to_csv("patientnoshownew.csv")
```
