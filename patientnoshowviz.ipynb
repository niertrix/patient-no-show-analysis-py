```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('patientnoshownew.csv')
df
```

```python
fig1 = px.box(df, x=(df['AppointmentDay'].astype('datetime64[ns]') - df['ScheduledDay'].astype('datetime64[ns]')).dt.days, labels={'x': 'Waiting Time', 'y': "Number of Patients"}, color='No-show', title='Does waiting time breed absence?')
fig1.show()
```

While some patients do not show up, the visuals show that most patients still show up regardless of waiting time.

```python
#Do text messages really matter? (No they don't, not even I read them, I just delete them all the time)
fig2 = px.histogram(df, x='SMS_received', y='No-show', title='Do text messages really matter?', labels={'SMS_received': 'Number of SMS messages received','No-show': 'Show'}, color='No-show') #Plot twist: NO ONE READS THEM!!
fig2.show()
```

Text messages do not make too much of a difference. Only about 25% of the people who recieve the SMS messages actually showed up.

```python
fig3 = px.histogram(df, x='Age', title='Age distribution of patients who show up vs those who do not show up', labels={'Age': 'Age of patient'}, color='No-show')
fig3.show()
```

The youth are surprisingly more reliable here. There is a larger difference between the number of patients who did not show up to the ones who did among the youth more than among the elderly
